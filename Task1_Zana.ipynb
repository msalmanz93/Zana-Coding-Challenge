#  **Sentence wise vector**

```
sentence1 = 'What are the main symptoms of diabetes?'
sentence2 = 'Tell me how I may treat bruxism'
sentence3 = 'What is the cause of a stroke?'
```

```
sv1 = ft.get_sentence_vector(sentence1)
sv2 = ft.get_sentence_vector(sentence2)
sv3 = ft.get_sentence_vector(sentence3)

print(sv1,sv2,sv3)
```

```
import pandas as pd
df = pd.DataFrame({'Sentence': [sentence1,sentence2,sentence3], 'Sentence Vector': [sv1,sv2,sv3]})
df.to_csv('sentenceVector.csv',index=False)

from google.colab import files
files.download("sentenceVector.csv")
```
